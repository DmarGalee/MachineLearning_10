```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from google.colab import files

# Upload Dataset
uploaded = files.upload()
df = pd.read_csv("wbc.csv")

# Cek Data
print("Informasi Data:")
print(df.info())
print("\nJumlah missing value per kolom:")
print(df.isnull().sum())

```

```python
# Hapus kolom yang tidak diperlukan
if 'id' in df.columns:
    df = df.drop('id', axis=1)

# Hapus kolom yang berisi semua NaN atau tidak ada variasi (constant)
df = df.dropna(axis=1, how='all')  # drop kolom semua NaN
df = df.loc[:, df.nunique() > 1]   # drop kolom dengan hanya 1 nilai unik

# Pisahkan X dan y
X = df.drop('diagnosis', axis=1)
y = df['diagnosis']

# Encoding Diagnosis
encoder = LabelEncoder()
y = encoder.fit_transform(y)
print("\nMapping Label:", dict(zip(encoder.classes_, encoder.transform(encoder.classes_))))

# Standarisasi
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Stratified Split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y,
    test_size=0.2,
    stratify=y,
    random_state=42
)

print("\nJumlah data latih:", X_train.shape[0])
print("Jumlah data uji:", X_test.shape[0])
print("\nDistribusi kelas pada data latih:")
print(pd.Series(y_train).value_counts(normalize=True))
print("\nDistribusi kelas pada data uji:")
print(pd.Series(y_test).value_counts(normalize=True))
```
