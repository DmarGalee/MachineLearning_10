```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.utils import to_categorical

# 1. Load dataset MNIST
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Normalisasi data (0-255 → 0-1)
X_train = X_train / 255.0
X_test = X_test / 255.0

# One-hot encoding label
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# 2. Bangun model JST
model = Sequential([
    Flatten(input_shape=(28, 28)),          # Ubah gambar 28x28 menjadi vektor
    Dense(128, activation='relu'),          # Hidden layer 1
    Dense(64, activation='relu'),           # Hidden layer 2
    Dense(10, activation='softmax')         # Output layer (10 kelas)
])

# 3. Kompilasi model
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# 4. Latih model
model.fit(X_train, y_train, epochs=5, batch_size=32)

# 5. Evaluasi model
loss, acc = model.evaluate(X_test, y_test)
print(f"Akurasi pada data uji: {acc:.4f}")
```

```python
## 256 dan 128 ,ReLU

import time
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.utils import to_categorical

# Load MNIST
(X_train, y_train), (X_test, y_test) = mnist.load_data()
X_train, X_test = X_train / 255.0, X_test / 255.0
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# Model 256–128
model = Sequential([
    Flatten(input_shape=(28,28)),
    Dense(256, activation='relu'),
    Dense(128, activation='relu'),
    Dense(10, activation='softmax')
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

start = time.time()
model.fit(X_train, y_train, epochs=5, batch_size=32)
train_time = time.time() - start

loss, acc = model.evaluate(X_test, y_test)
print("Akurasi:", acc)
print("Waktu pelatihan:", train_time, "detik")
```

```python
# Sigmoid
import time
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.utils import to_categorical

# Load MNIST
(X_train, y_train), (X_test, y_test) = mnist.load_data()
X_train, X_test = X_train / 255.0, X_test / 255.0
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# Model 256–128
model = Sequential([
    Flatten(input_shape=(28,28)),
    Dense(256, activation='sigmoid'),
    Dense(128, activation='sigmoid'),
    Dense(64, activation='sigmoid'),
    Dense(10, activation='softmax')
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

start = time.time()
model.fit(X_train, y_train, epochs=5, batch_size=32)
train_time = time.time() - start

loss, acc = model.evaluate(X_test, y_test)
print("Akurasi:", acc)
print("Waktu pelatihan:", train_time, "detik")
```

## Perbandingan Model Berdasarkan Hidden Layer, Aktivasi, Akurasi, dan Waktu Training

- Pada model dengan dua hidden layer berukuran 256 dan 128 neuron menggunakan aktivasi ReLU, model menunjukkan performa terbaik dan paling stabil untuk ukuran jaringan tersebut. Akurasinya mencapai 0.986, dengan waktu pelatihan sekitar 50–60 detik. ReLU terbukti cepat karena tidak mengalami masalah vanishing gradient.

- Ketika menggunakan ukuran hidden layer yang sama (256–128) tetapi aktivasi diganti menjadi Sigmoid, performa menurun. Akurasi yang diperoleh hanya berada pada rentang 0.965–0.973, dan waktu pelatihannya lebih lama, yaitu 55–60 detik. Penurunan kecepatan dan akurasi terjadi karena Sigmoid cenderung mengalami gradient vanishing pada jaringan yang lebih dalam.

- Pada model dengan tiga hidden layer (256–128–64) menggunakan ReLU, hasilnya menjadi sedikit lebih baik dibandingkan dua layer ReLU sebelumnya. Akurasinya mencapai 0.975, dengan waktu training sekitar 60–70 detik. Ini menjadi konfigurasi terbaik secara keseluruhan, meskipun waktu training bertambah.

- Sebaliknya, ketika jaringan tiga layer (256–128–64) menggunakan Sigmoid, akurasi hanya sekitar 0.96 dan waktu pelatihan semakin lambat, yaitu 60–70 detik. Kinerjanya cenderung stagnan karena jaringan yang lebih dalam memperparah masalah vanishing gradient pada Sigmoid.
