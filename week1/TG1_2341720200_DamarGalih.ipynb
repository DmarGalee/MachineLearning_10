Penjelasan Fungsi Masing-Masing Library

PyPREP

Kegunaan:
Library Python untuk Preprocessing EEG (Electroencephalography).
Fokus utamanya adalah implementasi algoritma PREP pipeline, yaitu metode otomatis untuk membersihkan sinyal EEG dari gangguan (misalnya line noise, bad channels, baseline drift).

Cocok untuk: penelitian EEG, BCI (Brain-Computer Interface), atau analisis sinyal otak.

SciPy

Kegunaan:
Salah satu library fundamental di Python untuk scientific computing.
Berisi banyak fungsi untuk aljabar linear, optimisasi, integrasi, transformasi Fourier, statistik, pemrosesan sinyal, dan analisis data ilmiah lainnya.

Cocok untuk: semua kebutuhan perhitungan matematis/ilmiah, terutama dalam penelitian data sains.

wandb (Weights & Biases)

Kegunaan:
Platform untuk tracking eksperimen machine learning.
Bisa digunakan untuk:

Logging metrik training (akurasi, loss, dsb) secara otomatis.

Visualisasi hasil training di dashboard online.

Menyimpan model, dataset, dan hasil eksperimen.

Cocok untuk: pengembangan Deep Learning/ML, kolaborasi tim, dan eksperimen terstruktur.

pyECG

Kegunaan:
Library Python untuk analisis sinyal ECG (Electrocardiography).
Fungsinya mencakup:

Deteksi R-peaks

Ekstraksi fitur HRV (Heart Rate Variability)

Preprocessing sinyal ECG (filtering, noise removal)

Cocok untuk: penelitian kesehatan jantung, wearable devices, bio-signal processing.

1. Tindakan Melanggar Etika dan Hukum dalam Penggunaan Kecerdasan Buatan
Contoh:

Deepfake untuk Penipuan atau Fitnah
AI digunakan untuk membuat video deepfake seseorang yang mengatakan atau melakukan sesuatu yang sebenarnya tidak pernah dilakukan. Ini bisa digunakan untuk penipuan finansial, manipulasi politik, atau pencemaran nama baik.

Melanggar Etika: karena merusak reputasi, menyebarkan misinformasi, dan mengancam kepercayaan publik.

Melanggar Hukum: di beberapa negara, penggunaan deepfake untuk tujuan kriminal bisa dipidana (misalnya terkait UU ITE di Indonesia yang mengatur penyebaran informasi palsu).

Referensi:

Chesney, R., & Citron, D. K. (2019). Deep Fakes: A Looming Challenge for Privacy, Democracy, and National Security. California Law Review.

UU No. 11 Tahun 2008 tentang Informasi dan Transaksi Elektronik (ITE) di Indonesia.

2. Dampak Energi dan Lingkungan dari Pemanfaatan Kecerdasan Buatan
Contoh Dampak:

Tingginya Konsumsi Energi dalam Pelatihan Model AI Besar
Melatih model AI skala besar (misalnya GPT, BERT, LLaMA) membutuhkan pusat data dengan ribuan GPU.

Studi dari University of Massachusetts Amherst (Strubell et al., 2019) menunjukkan bahwa melatih satu model NLP besar dapat menghasilkan emisi karbon setara dengan 5 mobil sepanjang masa pakainya.

Hal ini memperburuk jejak karbon dan meningkatkan beban lingkungan dari sektor teknologi.

Cara Mengatasinya (pendapat pribadi + solusi umum):

Optimisasi Model AI

Gunakan teknik efisiensi seperti model pruning, quantization, atau knowledge distillation untuk mengurangi kebutuhan komputasi.

Energi Terbarukan di Data Center

Perusahaan AI sebaiknya mengandalkan listrik dari sumber energi terbarukan (solar, wind, hydro).

Pemanfaatan Edge AI

Memindahkan sebagian pemrosesan ke perangkat lokal (edge devices) agar tidak selalu bergantung pada pusat data besar.

Kebijakan Green AI

Dorongan riset dan regulasi agar AI tidak hanya “powerful” tetapi juga ramah lingkungan.

Referensi:

Strubell, E., Ganesh, A., & McCallum, A. (2019). Energy and Policy Considerations for Deep Learning in NLP. ACL 2019.

Patterson, D. et al. (2021). Carbon Emissions and Large Neural Network Training. Google Research.
